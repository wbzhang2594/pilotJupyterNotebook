# Probability of meeting someone having same birthday with you.

这个问题是: 在多少个人之中,一定有一个人和你的生日在同一天.

换一种说法, 在多少个人之中, 不存在任何一个找不到和他有相同生日的人.

## Probability

The $P(n)$ denotes the probability of "there isn't anyone has same birthday with you", when there are number $n$ of other people as samples.

首先找到人数 $n$ 和 $P(n)$ 的关系.

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt 


x=range(0,10000)
p=np.power(364/365,x)   #probability
# print(type(x))
# print(type(p))
plt.plot(x,p)
plt.ylabel(r'$p(n)$')
# plt.ylabel("人数", fontproperties=zhfont1)
plt.xlabel('n')

plt.annotate(r'(' + str(x[0]) + r', ' +str(p[0]) + r')', xy=(x[0],p[0]), xycoords='data',xytext=(+10, +30), textcoords='offset points'
                 , arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.annotate(r'(' + str(x[1]) + r', ' +str(p[1]) + r')', xy=(x[1],p[1]), xycoords='data',xytext=(+10, -30), textcoords='offset points'
                 , arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.annotate(r'(' + str(x[1000]) + r', ' +str(p[1000]) + r')', xy=(x[1000],p[1000]), xycoords='data'
             ,xytext=(+10, +40), textcoords='offset points'
                 , arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.annotate(r'(' + str(x[2000]) + r', ' +str(p[2000]) + r')', xy=(x[2000],p[2000]), xycoords='data'
             ,xytext=(+10, +30), textcoords='offset points'
                 , arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.show()
```

```python
lg_p = np.log10(p)
```

```python
plt.xlabel('people num')
plt.ylabel('lg_p')
plt.plot(x,lg_p)

plt.show()
```

```python
np.array(x)[lg_p<-6]
# x1[isinstance(y1,int)][1]

```

```python
lg_p_special_points=np.arange(0,-12, -1)
lg_p_special_points
```

```python
x_special_points=[]
for item in lg_p_special_points:
    x_special_points.append(np.array(x)[lg_p<item][0])

x_special_points
```

```python
# zhfont1 = matplotlib.font_manager.FontProperties(fname="SourceHanSansSC-Bold.otf") 

def customizeLabel(number):
    if(number==0):
        return r'${'+ str(10^number) + r'}$'
    else:
        return r'$10^{'+ str(number) + r'}$'


p_ticks=[]
for i in range(0, lg_p_special_points.size):
    p_ticks.append(customizeLabel(lg_p_special_points[i]))
  

plt.yticks(lg_p_special_points, p_ticks)

plt.plot(x,lg_p)
plt.plot(x_special_points,lg_p_special_points,'x')
plt.ylabel(r'probability of "no one has same birthday"')
# plt.ylabel("人数", fontproperties=zhfont1)
plt.xlabel('people number')

for i in range(0, lg_p_special_points.size):
    tmp_lg_p_special_point=lg_p_special_points[i]
    tmp_x_special_point=x_special_points[i]
    plt.annotate(r'('+str(tmp_x_special_point) + r', $10^{'+ str(tmp_lg_p_special_point) + r'}$)'
                 ,xy=(tmp_x_special_point, tmp_lg_p_special_point), xycoords='data'
                 ,xytext=(+10, +30), textcoords='offset points'
                 , arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    
#     plt.plot([yy2,yy2],[0,xx2],color ='gray',  linewidth=1.5, linestyle="--")
#     plt.plot([0,yy2],[xx2,xx2],color ='gray',  linewidth=1.5, linestyle="--")
    
plt.show()
```
