```python
# coding:utf-8
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

D='date'
I='income'
E='expenditure'

with open('accounts.csv', newline='') as csvfile:
    dictReader = csv.DictReader(csvfile, delimiter=',', quotechar='|')
    dates=[]
    income=[]
    expenditure=[]
    for row in dictReader:
        dates.append(row[D])
        income.append(int(row[I]))
        expenditure.append(int(row[E]))

index=[i for i in range(len(dates))]
xs = [datetime.strptime(d, '%Y-%m-%d').date() for d in dates]
ys = range(len(xs))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())


plt.title("GT Recording") 
plt.xlabel("Date") 
plt.ylabel("GT in total") 
plt.plot(xs, income)
plt.gcf().autofmt_xdate() 
plt.show()



```
